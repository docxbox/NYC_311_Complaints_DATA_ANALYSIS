ANNOVA Test

```python
from scipy.stats import f_oneway

# Group Request_Closing_Time values by Complaint Type
groups = [group["Request_Closing_Time"].values for name, group in df.groupby("Complaint Type")]

# Run ANOVA test
anova_result = f_oneway(*groups)

# F-statistic and p-value
print("ANOVA Test Results:")
print("F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)

# result
if anova_result.pvalue < 0.05:
    print("Conclusion: Reject the null hypothesis At least one complaint type has a significantly different closing time.")
else:
    print("Conclusion: Fail to reject the null hypothesis No significant difference in closing time across complaint types.")
```

Chi-Square Test

```python
from scipy.stats import chi2_contingency

# Build contingency table
contingency = pd.crosstab(df['Complaint Type'], df['Borough'])

# Apply Chi-Square test
chi2, p_val, dof, expected = chi2_contingency(contingency)

# Display results
print("Test 2: Chi-Square Test")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print("\nObserved Frequencies:\n", contingency)
print("\nExpected Frequencies:\n", expected)

# Interpretation
if p_val < 0.05:
    print("Conclusion: Reject the null hypothesis. Complaint type is dependent on borough.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No evidence of a relationship.")
```
